```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

## Adding sample data

```python
# creating random data
features = 4 * np.random.rand(400, 1) - 2
label = 3 + 5 * features**2 + 2 * np.random.randn(400,1)

plt.scatter(features, label)
plt.show()
```

## Creating the model

- Splitting the data.
- Feeding the model.
- Predicting the test data.

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2)

model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print('Model Score:', model.score(X_test, y_test))

sorted_idx = np.argsort(X_test.flatten())
X_test_sorted = X_test[sorted_idx]
predictions_sorted = predictions[sorted_idx]

plt.scatter(X_train, y_train, alpha=0.6, label='Training Data')
plt.plot(X_test_sorted, predictions_sorted, color='r', label='Model Prediction', linewidth=3)
plt.title('Polynomial Regression')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```
